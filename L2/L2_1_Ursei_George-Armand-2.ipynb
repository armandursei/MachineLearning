```python
###
import os
import matplotlib.pyplot as plt
```

stabilirea caii catre date

```python
#calea catre directorul setului de date
dataset_folder = '/content/drive/MyDrive/Colab Notebooks/ds'
```

crearea strucutrii de tip dictionar pentru numararea fisierelor

```python
#creare dictionar
category_count = {}
```

parcurgerea setului de date si numararea imaginilor din fiecare subdirector


```python
#parcurgerea setului de date
for category in os.listdir(dataset_folder):
  category_path = os.path.join(dataset_folder, category)
  if os.path.isdir(category_path):
    category_count[category] = len(os.listdir(category_path))
#extragere etichete si numarare fiecare clasa numaratoare
category_names = list(category_count.keys())
print('Eticheta Clasa', category_names)
category_counts = list(category_count.values())
print('Index numarare Clasa', category_counts)
```

Afisare Grafica Distributie DataSet

```python
# Create a bar chart to visualize the distribution of categories
plt.figure(figsize = (12, 6))
plt.barh(category_names, category_counts)
plt.xlabel('Numar de Imagini')
plt.ylabel('Clase')
plt.title('Set de Date - Distributia Categoriilor de CLASE')
plt.gca().invert_yaxis() #Invert the y-axis to display the most common
plt.show()
```
