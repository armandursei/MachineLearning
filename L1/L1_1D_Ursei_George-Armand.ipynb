importarea bibliotecillor necesare

```python
###
import numpy as np
import pywt
from skimage.restoration import denoise_wavelet
import matplotlib.pyplot as plt
```

prelucrarea datelor

```python
# citire date din biblioteca Python integrata din EKG-uri
x = pywt.data.ecg().astype(float)/256 # semnalul ekg este citit si normalizat

#adaugare zgomot la semnalul initial ekg ( semnal de inalta frecventa)
sigma = 0.05 # factorul de varianta a zgomotului
x_noisy = x + sigma * np.random.randn(x.size) #adaugare zgomot la semnal

#eliminarea zgomotului din undina
x_denoise = denoise_wavelet(x_noisy, method='BayesShrink' , mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True')
```

afisarea rezultatelor

```python
#afisare/plotare SEMNALE (portocaliu-zgomot/albastru-corectat)
plt.figure(figsize=(20, 10), dpi=100)
plt.plot(x_noisy, color='orange') #portocaliu
plt.plot(x_denoise, color='blue') #albastru
plt.title('Lab1.1 * Eliminare zgomot EKG', fontsize=16)
plt.legend(['EKG Zgomot','EKG ok',])
plt.show()
```

```python
#afisare/plotare SEMNALE (verde-initial/albastru-corectat)
plt.figure(figsize=(20, 10), dpi=100)
plt.plot(x, color='green') #verde
plt.plot(x_denoise, color='blue') #albastru
plt.title('Lab1.1 * Eliminare zgomot EKG', fontsize=16)
plt.legend(['EKG initial','EKG ok',])
plt.show()
```
